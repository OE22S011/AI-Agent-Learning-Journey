```python
def ai_agent(ticket):
    print(f"\n📩 OBSERVE: New ticket received: '{ticket}'")
    
    if "payment" in ticket.lower():
        department="Payments Team"
    elif "login" in ticket.lower():
        department="Auth Team"
    elif "slow" in ticket.lower() or "performance" in ticket.lower():
        department="Performance Team"
    elif "refund" in ticket.lower():
        department="Refund Team"
    elif "error" in ticket.lower():
        department="Technical Team"
    else:
        department="General Support"
    
    print(f"🧠 THINK: This is is {department} issue")
    print(f"💫 ACT: Routing ticket to the {department}")
    print(f"✅ Done: Ticket Successfully Routed!\n")

ai_agent("My payment failed twice today")
ai_agent("I cannot login to my account")
ai_agent("The app is very slow")
ai_agent("I need help with my order")
ai_agent("I need the refund of my previous order")
ai_agent("I cant able to login, showing me an error")
```

```python
def smart_agent(ticket):
    print(f"📩 New Ticket Received: '{ticket}'")
    ticket_lower=ticket.lower()
    departments=[]
    
    if "payment" in ticket_lower:
        departments.append("Payments Team")
    if "login" in ticket_lower:
        departments.append("Auth Team")
    if "slow" in ticket_lower:
        departments.append("Perfromance Team")
    if "help" in ticket_lower:
        departments.append("General Support Team")
    if "refund" in ticket_lower:
        departments.append("Refunds Team")
    if "error" in ticket_lower:
        departments.append("Technical Team")
    if not departments:
        departments.append("General Team")
        
    print(f"🧠 THINK: This ticket involves: {departments}")
    if len(departments)==1:
        print(f"💫 ACT: Routing to {departments[0]}")
    else:
        print(f"💫 ACT: Multiple teams involved!")
        for dept in departments:
            print(f"  -> Notifying {dept}")
    print(f"✅ DONE: Ticket Successfully Routed!\n")

    
smart_agent("I cant able to login, showing me an error")
smart_agent("My payment failed and I want a refund")
smart_agent("App is slow and showing error on login")
smart_agent("I need help with my order")
        
```
