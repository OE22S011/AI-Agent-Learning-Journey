```python
from openai import AzureOpenAI
from dotenv import load_dotenv
from google.colab import userdata
import os

load_dotenv()

client=AzureOpenAI(
    api_key=userdata.get("AZURE_API_KEY"),
    azure_endpoint=userdata.get("AZURE_ENDPOINT"),
    api_version="2024-02-01"
)
print("❤️ Connected to Azure OpenAI successfully!")
```

```python
response = client.chat.completions.create(
    model="gpt4nano-deploy",
    messages=[
        {
              "role":"system",
              "content":"""You are a intelligent triage agent.
              Analyze the customer ticket and provide:
              1. Category of issue
              2. Priority (High/Medium/Low)
              3. Team to route to
              4. Brief reason"""
        },
        {
            "role":"user",
            "content": "My payment failed twice today and I want a refund",
            "content": "I cannot login since morning!"
            #"content": "App is extremely slow and crashing"
            #"content": "I was charged twice for same order"

        }
    ]
)
print("🤖 AI Triage Agent Response: ")
print(response.choices[0].message.content)
```

```python
def ai_triage_agent(ticket):
  print(f"\n📩 New Ticket : '{ticket}'")
  print("-"*50)
  response=client.chat.completions.create(
      model="gpt4nano-deploy",
      messages=[
          {
              "role":"system",
              "content":"""You are an intelligent triage agent.
              Analyze the customer ticket and provide:
              1. Category of issue
              2. Priority (High/Medium/Low)
              3. Team to route to
              4. Brief reason"""
          },
          {
              "role":"user",
              "content":ticket
          }
      ],
      max_completion_tokens=200
  )
  print("🤖AI Response:")
  print(response.choices[0].message.content)
  print("-"*50)

tickets=[
    "My payment failed twice today!",
    "I cannot login since morning!",
    "App is extremely slow and crashing!",
    "I was charged twice for same order!",
    "I need help with my account settings!"
]
for ticket in tickets:
  ai_triage_agent(ticket)
```
